- thalach : maximum heart rate achieved

- target : 0= less chance of heart attack 1= more chance of heart attack

```python
X = heart_df.drop("output",axis=1)
y = heart_df['output']
```

```python
y.value_counts(normalize=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
def build_pipeline(model):
    
    pipeline = Pipeline([
                       ('scaler', StandardScaler()),
                       ('clf', model) 
                       ])
    
    return pipeline
```

```python
def scan_model(pipeline,param_dict,X_train,y_train):
    
    grid_cv = GridSearchCV(pipeline,param_grid=param_dict,scoring='accuracy',cv=5,verbose=2)
    grid_cv.fit(X_train,y_train)
    
    #print("The best parameter combination : {}".format(grid_cv.best_params_))
    #print("The best score on the training set : {}".format(grid_cv.best_score_))
    
    return grid_cv
```

```python
def eval_model(model,X_test,y_test):
    
#    pred_train = model.predict(X_train)
#    tn,fp,fn,tp = confusion_matrix(y_train,pred_train)
#    acc_train = accuracy_score(y_train,pred_train)
#    prec_score = precision_score(y_train,pred_train)
#    recall_Score = recall_score(y_train,pred_train)
    
    pred_test = model.predict(X_test)
#    tn,fp,fn,tp = confusion_matrix(y_test,pred_test)
    acc_test = accuracy_score(y_test,pred_test)
    prec_test = precision_score(y_test,pred_test)
    recall_test = recall_score(y_test,pred_test)
    f1_test = f1_score(y_test,pred_test)
    
#    print("confusion matrix of the train set : ")
#    print(confmat_train)
#    print("\n")
#    print("classificatio report of the train set : ")
#    print(report_train)
#    print("\n")
    
#    print("confusion matrix of the test set : ")
#    print(confmat_test)
#    print("\n")
#    print("classificatio report of the test set : ")
#    print(report_test)
#    print("\n")

    return acc_test,f1_test#,tn,fp,fn,tp
```

```python
model_list = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier(),
        GradientBoostingClassifier(), AdaBoostClassifier()]

lr_dict = {'clf__C' : np.arange(0.1,6,0.5),
           'clf__solver': ['liblinear','lbfgs']}

dt_dict = {'clf__criterion' : ['gini', 'entropy'],
           'clf__max_features' : ['sqrt','log2'],
           'clf__max_depth' : np.arange(10,60,10)}

rf_dict = {'clf__n_estimators': np.arange(20,100,20),
           'clf__criterion' : ['gini', 'entropy'],
           'clf__max_features' : ['sqrt','log2'],
           'clf__max_depth' : np.arange(10,60,10)}

knn_dict = {'clf__n_neighbors': np.arange(5,30,5),
            'clf__weights' : ['uniform','distance'],
            'clf__algorithm' : ['auto','ball_tree','kd_tree','brute'],
            'clf__metric' : ['minkowski','euclidean']}

gb_dict = {'clf__loss' : ['deviance','exponential'],
           'clf__learning_rate': np.arange(0.1,5,0.5),
           'clf__n_estimators': np.arange(20,100,20),
           'clf__criterion' : ['friedman_mse','mse','mae'],
           'clf__max_features' : ['sqrt','log2']}

ada_dict = {
           'clf__learning_rate': np.arange(0.1,5,0.5),
           'clf__n_estimators': np.arange(20,100,20),
           'clf__algorithm' : ['SAMME','SAMME.R']}

param_list = [lr_dict, dt_dict, rf_dict, knn_dict, gb_dict, ada_dict]
```

```python
scan_list = []
pred_list = []

for model, param_dict in zip(model_list,param_list):
    print("building model : {}\n".format(str(model)[:-2]))
    pipeline = build_pipeline(model)
    start_time = time.time()      
    grid_cv = scan_model(pipeline,param_dict,X_train,y_train)
    print("\n model fitted!")
    elapsed_time = time.time() - start_time        
    acc_test,f1_test = eval_model(grid_cv,X_test,y_test)   # ,tn,fp,fn,tp
      
    scan_list.append({
              'model': str(model)[:-2],
              'best_score': grid_cv.best_score_,
              'test_acc': acc_test,
              'test_f1': f1_test,
              'best_params':grid_cv.best_params_,
              'time':elapsed_time
    })      
          
          
#    pred_list.append({
#              'model': str(model)[:-2],
#              'true_neg':tn,
#              'false_pos':fp,
#              'false_neg':fn,
#              'true_pos':tp
#    })      
```
